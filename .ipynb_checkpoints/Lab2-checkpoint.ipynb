допомогою простого випадкового відбору без повернення.

```r
file = read.table('StatVillage.raw', colClasses = c("V4"="character"))
names = c('BLOCK', 'UNIT', 'HHSIZE', 'HHPER', 'NUEMPINH', 'NUIRH', 'EMPINCH', 'INVSTH', 'GOVINCH', 'OTINCH', 'TOTINCH', 'DTYPEH', 'BUILTH', 'TENURH', 'MORGH', 'ROOMH', 'BROOMH', 'VALUEH', 'GROSRTH', 'OMPH', 'HMAGE', 'HMSEX', 'HMMTN', 'HMHLOS', 'HMOCC91', 'HMLFACT', 'HMWKSWK', 'HMEMPIN', 'SHMAGE', 'SHMSEX', 'SHMMTN', 'SHMHLOS', 'SHMOCC91', 'SHMLFACT', 'SHMWKSW', 'SHMEMPIN')
colnames(file) = names
```

```r
set.seed(0)
N <- length(file[,'BLOCK'])
x <- sample(1:1023,N/4)
data <- file[x,]
```

```r
data2 <- data[,c('GOVINCH','HHSIZE')]
data2['GOVINCHPH'] <- data2[,'GOVINCH']/data2[,'HHSIZE']
```

```r
head(data2)
```

Оцінити параметр R. 

```r
GOVINCH_sum <- sum(data2[,'GOVINCH'])
HHSIZE_sum <- sum(data2[,'HHSIZE'])
R <- GOVINCH_sum/HHSIZE_sum
R
```

## Завдання 2 - 6 балів
Оцінити дисперсію оцінки параметра R трьома методами: за допомогою лінеаризації
Тейлора, Джек-найф та Бут-стреп на основі однієї і тієї ж вибірки.

## Лінеризація Тейлора

Використаємо другу формулу для дисперсії

```r
N <- 1024
n <- 1024/4
f <- n/N
y <- GOVINCH_sum
z <- HHSIZE_sum
R <- GOVINCH_sum/HHSIZE_sum 
s_z <- var(data2[,'HHSIZE'])
s_y <- var(data2[,'GOVINCH'])
cov <- cov(data2[,'GOVINCH'],data2[,'HHSIZE'])
D <- (s_y+R*R*s_z-2*R*cov)/(z*z)
D 
```

```r
head(data2)
```

```r
n <- 1024/4
sigma <- R
y_sum <- GOVINCH_sum
z_sum <- HHSIZE_sum
var_jk2 <-0
sigma_jk <- 0
var_jk1 <- 0 
for (i in 1:nrow(data2)){
    sigma_j <- (y_sum-data2[i,'GOVINCH'])/(z_sum-data2[i,'HHSIZE'])
    var_jk2 = var_jk2 + (sigma_j-sigma)^2
    sigma_jk = sigma_jk + (y_sum-data2[i,'GOVINCH'])/(z_sum-data2[i,'HHSIZE'])
}
sigma_jk = sigma_jk/n
for (i in 1:nrow(data2)){
    sigma_j <- (y_sum-data2[i,'GOVINCH'])/(z_sum-data2[i,'HHSIZE'])
    var_jk1 = var_jk1 + (sigma_j-sigma_jk)^2
}
var_jk2 = var_jk2*(n-1)/n

var_jk1 = var_jk1*(n-1)/n
print(var_jk2)
print(var_jk1)
```

$D(JK_1) = 19189.47$

## Бут страп


Для того, щоб зробити бутстрапову вибірку застосуємо алгоритм,який не був описаний в лекції, а саме зробимо так. Сформуємо вибірку роблячи звичайний ПВВПП розміру 16 з нашої підвибірки. 
Повторимо цю операцію 64 рази -> у нас буде вибірка розміру 1024, скільки нам і треба для генеральної сукупності

```r
16*64
```

```r
x <- sample(1:256,16)
data_boot <- data2[x,]
for(i in 1:63){
    x <- sample(1:256,16)
    data_boot <- rbind(data_boot,data2[x,])
}
```

```r
data_all <- rbind(data2,data2,data2,data2)
number <- 1000
boot_list <- list()
for (i in 1:number){
    x <- sample(1:1024,256)
    calculate_data <- data_all[x,]
    app = sum(calculate_data[,'GOVINCH'])/sum(calculate_data[,'HHSIZE'])
    boot_list[i] <- app
}
boot_list2 <- unlist(boot_list, use.names=FALSE)
sigma_mean <- mean(boot_list2)
var_sigma <- var(boot_list2)
print(sigma_mean)
print(var_sigma)
```
